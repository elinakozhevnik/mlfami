## Коротко о теории

### 1. Логистическая регрессия
Логистическая регрессия моделирует вероятность положительного класса:

\[
P(y=1 \mid x) = \sigma(z), \quad z = w^T x + b
\]

где \(\sigma(z)\) — сигмоида:

\[
\sigma(z) = \frac{1}{1 + e^{-z}}
\]

### 2. Почему не линейная регрессия
Если предсказывать класс линейной моделью напрямую, прогноз может выйти за диапазон \([0, 1]\).  
Сигмоида решает эту проблему: она переводит любое число в вероятность.

### 3. Функция потерь
Для логистической регрессии используется **log-loss**:

\[
L = - \frac{1}{n} \sum_{i=1}^{n}
\left[
y_i \log(\hat{p}_i) + (1-y_i)\log(1-\hat{p}_i)
\right]
\]

### 4. Обучение
Мы будем минимизировать log-loss методом **градиентного спуска**.

### 5. Что будем сравнивать
После своей реализации вы сравните результат с `sklearn.linear_model.LogisticRegression`:
- accuracy,

### 6. Почему нужна стандартизация
Градиентный спуск работает стабильнее, когда признаки находятся примерно в одном масштабе.

## Шаг 2. Реализуем сигмоиду

Допишите функцию `sigmoid(z)`.

Проверка:
- `sigmoid(0)` должно быть равно `0.5`;
- функция должна работать и со скалярами, и с `numpy`-массивами.

```python
def sigmoid(z):
    # YOUR CODE HERE
    return 1/ (1 + np.exp(-z))

print("sigmoid(0) =", sigmoid(0))
print("sigmoid([-1, 0, 1]) =", sigmoid(np.array([-1.0, 0.0, 1.0])))
```

## Шаг 3. Реализуем log-loss

Напишите функцию `compute_log_loss(y_true, y_pred_proba)`.

Важно:
- чтобы избежать `log(0)`, используйте `np.clip(y_pred_proba, 1e-15, 1 - 1e-15)`.

```python
def compute_log_loss(y_true, y_pred_proba):
    y_pred_proba = np.clip(y_pred_proba, 1e-15, 1-1e-15)
    loss = -np.mean(
        np.log(y_pred_proba) * y_true + (1-y_true) * np.log(1-y_pred_proba)
    )

    return loss

toy_y = np.array([0, 1, 1, 0])
toy_p = np.array([0.1, 0.9, 0.8, 0.3])
print("Toy log-loss:", compute_log_loss(toy_y, toy_p))
```

## Шаг 4. Один шаг градиентного спуска

Напомним:
- \(z = Xw + b\)
- \(\hat{p} = \sigma(z)\)

Градиенты для log-loss:
\[
dw = 
rac{1}{n} X^T(\hat{p} - y), \qquad db = 
rac{1}{n}\sum(\hat{p} - y)
\]

Ниже нужно реализовать функцию, которая считает:
- вероятности,
- loss,
- градиенты `dw`, `db`.

```python
def forward_backward(X, y, w, b):
    z = X@w +b
    p = 1 / (1 + np.exp(-z))
    loss = compute_log_loss(y, p)
    dw = (X.T @ (p-y)) / X.shape[0]
    db = np.mean(p-y)

    # 1) посчитать z
    # 2) посчитать p = sigmoid(z)
    # 3) посчитать loss
    # 4) посчитать dw и db - изменение весов, функция forward - ф обновления весов
    ...
    return p, loss, dw, db

n_features = X_train_scaled.shape[1]
w0 = np.zeros(n_features)
b0 = 0.0

p0, loss0, dw0, db0 = forward_backward(X_train_scaled, y_train.values, w0, b0)
print("Initial loss:", loss0)
print("dw shape:", dw0.shape)
print("db:", db0)
```

## Шаг 5. Собираем модель в класс

Допишите методы:
- `fit`
- `predict_proba`
- `predict`

Требования:
- обучение через градиентный спуск;
- сохранять историю значения loss;
- в `predict` использовать порог `threshold`.

```python
class MyLogisticRegressionGD:
    def __init__(self, learning_rate=0.05, n_iters=3000, threshold=0.5):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.threshold = threshold
        self.w = None
        self.b = None
        self.loss_history = []

    def fit(self, X, y):
        n_features = X.shape[1]
        self.w = np.zeros(n_features)
        self.b = 0.0
        self.loss_history = []   

        for _ in range(self.n_iters):
            p, loss, dw, db = forward_backward(X, y, self.w, self.b)

            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

            self.loss_history.append(loss)
        # YOUR CODE HERE
        # 1) инициализируйте w и b
        # 2) в цикле вызовите forward_backward
        # 3) обновите параметры
        # 4) сохраните loss в self.loss_history
        ...
        return self

    def predict_proba(self, X):
        z = X @ self.w + self.b
        # пропускаем через сигмоиду для получения вероятностей
        proba = 1 / (1 + np.exp(-z))

        # YOUR CODE HERE
        return proba

    def predict(self, X):
        proba = self.predict_proba(X)
        preds = (proba >= self.threshold).astype(int)
        return preds
```

1.сигмоида переводит любое вещественное число в диапазон (0,1). это важно, так как выход модели мы интерпретируем как вероятность. без сигмоиды мы бы получали значения вроде −500 или +1000, которые невозможно использовать для классификации

2.линия графика MSE + sigmoid выравнивается при больших значениях ошибки , получается поверхность с множеством локальных минимумов. градиентный спуск может в них застрять и решить, что минимум достигнут, хотя ошибка всё еще огромна. У Log-loss + sigmoid только один глобальный минимум и в любой точке градиент четко указывает направление к нему. даже если модель сильно ошибается, наклон функции очень крутой, что заставляет алгоритм быстро исправлять веса

3.если признаки имеют разные масштабы (от 0 до 1 и от 0 до 100), поверхность функции потерь будет сильно вытянутой. стандартизация делает поверхность более круглой, позволяя градиентому спуску идти к минимуму кратчайшим путем

4.accuracy моей модели достигла 0.9737, в то время как библиотечная модель показала 0.9825. 
ROC-AUC: моя 0.9957, что даже несколько выше результата scikit-learn 0.9954. 
почти идентичные значения Precision и Recall подтверждают, что мой алгоритм градиентного спуска имеет высокую точность реализации функции потерь и градиентов, практически не уступая библиотечному

5.в scikit-learn используются оптимизированные на языке C солверы, которые умеют менять шаг обучения. также там  встроена регуляризация, предотвращающая переобучение, и более сложные критерии остановки цикла

6.при снижении порога мы чаще ставим диагноз; это позволяет найти всех больных (recall растет), но мы больше ошибаемся и записываем в больные даже здоровых (precision падает).
при поднятии порога модель становится очень осторожной и говорит 100% только когда максимально уверена. precision растет, но из-за такой придирчивости мы начинаем пропускать реальные случаи болезни

7.она идеальна для задач, где нужно понимать вес каждого признака, и где зависимость линейная. уступает более сложным моделям (случайный лес, нейросети) в задачах со сложными, нелинейными связями.
